```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit

# === Leer datos ===
data = np.genfromtxt("data.txt", skip_header=1, missing_values='--', filling_values=np.nan)
ngulos = data[:, 4]
intensidades = data[:, 5]

# Filtrar valores vlidos
valid = ~np.isnan(ngulos) & ~np.isnan(intensidades)
ngulos = ngulos[valid]
intensidades = intensidades[valid]

# === Modelo de Ley de Malus ===
def malus(, I0, 0, fondo):
    return I0 * np.cos(np.deg2rad( - 0))**2 + fondo

# Ajuste con incertidumbre
params, cov = curve_fit(
    malus, ngulos, intensidades,
    p0=[max(intensidades), 0, min(intensidades)],
    bounds=([0, -180, 0], [np.inf, 180, np.inf])
)
I0, 0, fondo = params
_I0, _0, _fondo = np.sqrt(np.diag(cov))  # Incertidumbres

ngulos_fit = np.linspace(0, 360, 500)
intensidades_fit = malus(ngulos_fit, I0, 0, fondo)

# === Estilo general para publicacin ===
plt.rcParams.update({
    "font.size": 10,
    "font.family": "serif",
    "axes.labelsize": 11,
    "axes.titlesize": 12,
    "legend.fontsize": 9,
    "xtick.labelsize": 9,
    "ytick.labelsize": 9,
    "axes.linewidth": 0.8,
    "xtick.direction": "out",
    "ytick.direction": "out",
    "xtick.major.size": 4,
    "ytick.major.size": 4,
    "figure.dpi": 300,
})

# === Grfica Cartesiana ===
fig1 = plt.figure(figsize=(6, 5))
ax1 = fig1.add_subplot(111)
ax1.plot(ngulos, intensidades, 'o', label="Datos", color="#4d4d4d", markersize=4)
ax1.plot(ngulos_fit, intensidades_fit, '-', label="Ajuste Malus", color="#d62728", linewidth=1)
ax1.set_xlabel("ngulo  ()")
ax1.set_ylabel("Intensidad I (lux)")
ax1.grid(True, linestyle="--", linewidth=0.5, alpha=0.5)
ax1.legend(loc='upper right', bbox_to_anchor=(1, 1), borderaxespad=0, frameon=True, edgecolor='black')
plt.tight_layout()

# === Grfica Polar ===
fig2 = plt.figure(figsize=(6, 5))
ax2 = fig2.add_subplot(111, polar=True)
ng_rad = np.deg2rad(ngulos)
ng_fit_rad = np.deg2rad(ngulos_fit)

ax2.plot(ng_rad, intensidades, 'o', label="Datos", color="#4d4d4d", markersize=4)
ax2.plot(ng_fit_rad, intensidades_fit, '-', label="Ajuste Malus", color="#d62728", linewidth=1)

# === Elipse inscrita ===
a = I0 + fondo
b = fondo
 = np.deg2rad(0)
t = np.linspace(0, 2*np.pi, 400)
x = a * np.cos(t)
y = b * np.sin(t)
x_rot = x * np.cos() - y * np.sin()
y_rot = x * np.sin() + y * np.cos()
r_elipse = np.sqrt(x_rot**2 + y_rot**2)
_elipse = np.arctan2(y_rot, x_rot)
ax2.plot(_elipse, r_elipse, '--', color='black', linewidth=1,
         label=f"Elipse inscrita\n = {0:.1f}  {_0:.1f}")

ax2.set_rticks(np.linspace(0, I0*1.1, 5))
ax2.grid(True, linestyle="--", linewidth=0.5, alpha=0.5)
ax2.legend(loc='upper left', bbox_to_anchor=(1, 1), borderaxespad=0, frameon=True, edgecolor='black')
plt.tight_layout()
fig2.savefig("polarizacion_linterna.pdf", dpi=300, bbox_inches='tight')


plt.show()

# === Resultados ===
print(f"I (mxima): {I0:.2f}  {_I0:.2f} lux")
print(f": {0:.2f}  {_0:.2f}")
print(f"Fondo (mnimo): {fondo:.2f}  {_fondo:.2f} lux")
print(f"Excentricidad: {np.sqrt(1 - (b / a)**2):.3f}")
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit

# Cargar los datos desde el archivo
data = np.genfromtxt("data.txt", skip_header=1, missing_values='--', filling_values=np.nan)

# Columnas:
# Ley de Malus con sensor de intensidad
# Primer polarizador horizontal
H = data[:, 0]
IH = data[:, 1]
# Primer polarizador vertical
V = data[:, 2]
IV = data[:, 3]

# Filtrar datos vlidos (sin NaN)
valid_H = ~np.isnan(H) & ~np.isnan(IH)
H = H[valid_H]
IH = IH[valid_H]

valid_V = ~np.isnan(V) & ~np.isnan(IV)
V = V[valid_V]
IV = IV[valid_V]

# Funcin modelo (Ley de Malus con fondo constante)
def malus(, I0, 0, fondo):
    rad = np.deg2rad( - 0)
    return I0 * np.cos(rad)**2 + fondo

# Lmites fsicos: I0  0, fondo  0
bounds = ([0, -180, 0], [np.inf, 180, np.inf])

# Ajuste horizontal
popt_H, _ = curve_fit(malus, H, IH, p0=[max(IH), 0, min(IH)], bounds=bounds)
I0_H, 0_H, fondo_H = popt_H
H_fit = np.linspace(min(H), max(H), 500)
IH_fit = malus(H_fit, *popt_H)

# Ajuste vertical
popt_V, _ = curve_fit(malus, V, IV, p0=[max(IV), 0, min(IV)], bounds=bounds)
I0_V, 0_V, fondo_V = popt_V
V_fit = np.linspace(min(V), max(V), 500)
IV_fit = malus(V_fit, *popt_V)

# Graficar
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.plot(H, IH, 'o', label='Datos H')
plt.plot(H_fit, IH_fit, '-', label='Ajuste H')
plt.xlabel("ngulo  ()")
plt.ylabel("Intensidad (lux)")
plt.title("Polarizador Horizontal")
plt.grid(True)
plt.legend()

plt.subplot(1, 2, 2)
plt.plot(V, IV, 'o', label='Datos V')
plt.plot(V_fit, IV_fit, '-', label='Ajuste V')
plt.xlabel("ngulo  ()")
plt.ylabel("Intensidad (lux)")
plt.title("Polarizador Vertical")
plt.grid(True)
plt.legend()

plt.suptitle("Ley de Malus  Ajustes con fondo fsico")
plt.tight_layout()
plt.show()

# Mostrar parmetros
print("== Polarizador Horizontal ==")
print(f"I: {I0_H:.2f} lux")
print(f": {0_H:.2f}")
print(f"Fondo: {fondo_H:.2f} lux\n")

print("== Polarizador Vertical ==")
print(f"I: {I0_V:.2f} lux")
print(f": {0_V:.2f}")
print(f"Fondo: {fondo_V:.2f} lux")
```
